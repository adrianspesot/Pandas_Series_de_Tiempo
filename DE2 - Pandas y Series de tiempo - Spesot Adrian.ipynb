```python
#Librerías Utilizadas
import numpy as np
import pandas as pd
```

#### Lectura del archivo Summer-Olympic-medals-1976-to-2008.csv

```python
# Realizo la lectura del archivo y lo deposito en la varialbe a
a = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv',encoding='latin-1')

# Como se aprecia que en el .csv hay muchas filas intermedias con valores NaA
#Procedo a eliminar las filas completas con NaN
archivo = a.dropna()

#Visualizo el archivo limpio
archivo
```

## 1 - Con este archivo, construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos

```python
#Objeto de series con cuyos índices son los años
# Como los datos de la colúmna Year son float y los necesito como int, realizo la conversión con astype('int64')
indice = pd.Series(archivo['Year'].astype('int64')).drop_duplicates()

# Con drop_duplicate() busco eliminar los años duplicados, si no era el objetivo solo es quitar el ultimo metodo
# Imprimo la serie de tiempo con los datos de la colúmna Year
indice

```

## 2 - Separar nombres de apellidos en dos columnas distintas usando la coma como separador
 

```python
# Crear la serie con los nombres de los atletas
nombres = pd.Series(archivo['Athlete'])

#Separar Nombres y apellidos en 2 colúmnas
nombres.str.split(',',n=1,expand=True) #n=1 significa en cuantos pasos lo haremos, si agrego mas pasos crea columnas tipo none ya que no encuentra datos
```

## 3 - Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

```python
# Creo el DataFrame de archivo a usar que se ha realizado la lectura como archivo
df_archivo = pd.DataFrame(archivo)
df_archivo
```

### Obtener el pais que mas medallas obtuvo

```python
# 1 - Creo un dataframedonde obtenga el nombre del país y la cantidad de medallas obtenidas
Pais_mas_medallero = pd.DataFrame(df_archivo.groupby('Country')['Medal'].count())

# 2 - Ejecuto metodo sort_values() con el fin de ordenar el df de manera descendente
# 3 - Agrego el método head() con agrumento 1 para decirle que me muestre la primer fila del df
# Como el df Pais_mas_medallero está ordenado me mostrará el que más medallas obtuvo

Pais_mas_medallero.sort_values('Medal',ascending=False).head(1)
```
